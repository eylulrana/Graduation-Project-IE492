### Merging Demographic and Spending Data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# load and merge files
transaction_data = pd.read_excel("kartdata1.xlsx")
demographical_data = pd.read_excel("demografik.xlsx")

df_fs = pd.merge(transaction_data, demographical_data, on="Customer_id", how="left")
df_fs.to_excel("df_fs.xlsx", index=False)
```

### Adding Slope, MA and Churn Columns to the Data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df_fs = pd.read_excel("df_fs.xlsx")
slope_df = pd.read_csv("slope.csv")
MA_df = pd.read_csv("MA.csv")
df_fs["trxdate"] = pd.to_datetime(df_fs["trxdate"]).dt.strftime('%Y-%m')
churn_df = pd.read_csv("churn_analysis.csv")
merged_df = pd.merge(df_fs, churn_df, how="left", on=["Customer_id", "trxdate"])
merged_df = pd.merge(merged_df, slope_df, how="left", on=["Customer_id", "trxdate"])
merged_df = pd.merge(merged_df, MA_df, how="left", on=["Customer_id", "trxdate"])
merged_df.head()
```

### Changing Null Values of Spending Categories with 0

```python
MCCPI = [
    "total_amount_market",
    "total_qty_count_market",
    "total_amount_diger",
    "total_qty_count_diger",
    "total_amount_giyim",
    "total_qty_count_giyim",
    "total_amount_saglik_kozmetik",
    "total_qty_count_saglik_kozmetik",
    "total_amount_sigorta",
    "total_qty_count_sigorta",
    "total_amount_teknoloji",
    "total_qty_count_teknoloji",
    "total_amount_ulasim",
    "total_qty_count_ulasim",
    "total_amount_yemek",
    "total_qty_count_yemek"
]

merged_df[MCCPI] = merged_df[MCCPI].fillna(0)


print(merged_df.isnull().sum())
```

### Deleting Unnamed Column

```python
merged_df = merged_df.drop(columns=merged_df.columns[merged_df.columns.str.contains('Unnamed')])
print(merged_df.head())
```

### Handling the Null Values

```python
import pandas as pd

null_marital_row = merged_df[merged_df['MaritalStatus'].isnull()]

boş_customer_ids = null_marital_row['Customer_id']

merged_df = merged_df[~merged_df['Customer_id'].isin(boş_customer_ids)]

silinen_satır_sayısı = len(null_marital_row)
print("Number of Deleted Rows:", silinen_satır_sayısı)
```

### Handling Null Values of EmployerSector

```python
import pandas as pd

#Replacing some EmployerSector values with "Sektör Yok"
condition = merged_df['EmploymentType'].isin(['Çalışmıyor', 'Emekli Çalışan', 'Serbest Meslek Sahibi'])
merged_df.loc[condition, 'EmployerSector'] = 'Sektör Yok'
 
# Delete Null values whose Employment type is Ücretli
isveren_df = merged_df[(merged_df['EmploymentType'] == 'Ücretli') & (merged_df['EmployerSector'].isnull())]
for customer_id in isveren_df['Customer_id'].unique():
    rows_to_drop = merged_df[merged_df['Customer_id'] == customer_id]
    merged_df.drop(rows_to_drop.index, inplace=True)

#For "İşveren" EmploymentType, set EmployerSector as "Özel Kurum" 
isveren_indexes = merged_df.index[merged_df['EmploymentType'] == 'İşveren']
merged_df.loc[isveren_indexes, 'EmployerSector'] = 'Özel Kurum'

#For 'Tacir/Esnaf' EmploymentType, set EmployerSector as "Özel Kurum" 
esnaf_indexes = merged_df.index[merged_df['EmploymentType'] == 'Tacir/Esnaf']
merged_df.loc[esnaf_indexes, 'EmployerSector'] = 'Özel Kurum'
 
# Filling the rest of null values as "Sektör Yok"
merged_df['EmployerSector'].fillna('Sektör Yok', inplace=True)
 
print(merged_df.isnull().sum())
 
unique_customers = merged_df[merged_df['EmployerSector'].isnull()]['Customer_id'].unique()
for customer_id in unique_customers:
    Employment_type = merged_df.loc[merged_df['Customer_id'] == customer_id, 'EmploymentType'].iloc[0]
    print(f"Customer ID: {customer_id}, EmploymentType: {Employment_type}")
```

### Filling Null Values of City with the City information of BranchName Column

```python
import pandas as pd

cities_of_branch = pd.read_excel("CitiesofBranch.xlsx")

null_city_rows = merged_df[merged_df['City'].isnull()]

for index, row in null_city_rows.iterrows():
    branch_name = row['BranchName']
    
    city_name = cities_of_branch.loc[cities_of_branch['branchname'] == branch_name, 'CityName'].values
    
    if len(city_name) > 0:
        merged_df.at[index, 'City'] = city_name[0]

print(merged_df.head(10))
```

### Deleting Unnecessary Columns

### Handling Null Values of EmploymentType

Understand Which Profession Codes has Null EmploymentTypes

```python
import pandas as pd
 
unique_Employment_types = merged_df["EmploymentType"].unique()
 
null_employment_df = merged_df[merged_df['EmploymentType'].isnull()]["ProfessionCode"]
unique_professionCode = null_employment_df.unique()

print(unique_Employment_types)
print(unique_professionCode)
```

Assigning Appropriate EmploymentTypes for Different ProfessionCodes

```python

null_employment_df = merged_df[(merged_df['EmploymentType'].isnull()) & 
                               (merged_df['ProfessionCode'].isin([17, 83, 32, 30, 39, 38, 25, 24, 60, 55, 34, 50]))]
 
for index, row in null_employment_df.iterrows():
    profession_code = row['ProfessionCode']
    if profession_code in [17, 38]:
        merged_df.at[index, 'EmploymentType'] = 'Çalışmıyor'
    elif profession_code in [25, 83]:
        merged_df.at[index, 'EmploymentType'] = 'Tacir/Esnaf'
    else:
        merged_df.at[index, 'EmploymentType'] = 'Ücretli'
```

### Changing Recording Time to Tenure

```python
import pandas as pd

today_date = pd.to_datetime('today').strftime('%Y-%m-%d')

merged_df['RecordingTime'] = pd.to_datetime(merged_df['RecordingTime']).dt.strftime('%Y-%m-%d')

merged_df['Tenure'] = (pd.to_datetime(today_date) - pd.to_datetime(merged_df['RecordingTime'])).dt.days / 365
merged_df.drop(columns=['RecordingTime'], inplace=True)
```

### Encoding the ValueSegment

```python
ValueSegment_encoded = pd.get_dummies(merged_df['ValueSegment'], prefix='ValueSegment')

ValueSegment_encoded = ValueSegment_encoded.astype(int)

merged_df = pd.concat([merged_df, ValueSegment_encoded], axis=1)

merged_df.drop('ValueSegment', axis=1, inplace=True)

print(merged_df.head())
```

### Encoding the EmploymentType

```python
EmploymentType_encoded = pd.get_dummies(merged_df['EmploymentType'], prefix='EmploymentType')

EmploymentType_encoded = EmploymentType_encoded.astype(int)

merged_df = pd.concat([merged_df, EmploymentType_encoded], axis=1)

merged_df.drop('EmploymentType', axis=1, inplace=True)

print(merged_df.head())
```

### Encoding the EmployerSector

```python
EmployerSector_encoded = pd.get_dummies(merged_df['EmployerSector'], prefix='EmployerSector')

EmployerSector_encoded = EmployerSector_encoded.astype(int)

merged_df = pd.concat([merged_df, EmployerSector_encoded], axis=1)

merged_df.drop('EmployerSector', axis=1, inplace=True)

print(merged_df.head(2))
print(merged_df.dtypes)
```

### Changing City with Geographical Region

```python
geographical_regions = {
    'Marmara': ['Edirne', 'Kırklareli', 'Tekirdağ', 'İstanbul', 'Kocaeli', 'Yalova', 'Sakarya', 'Bilecik', 'Bursa', 'Balıkesir', 'Çanakkale'],
    'İç Anadolu': ['Aksaray', 'Ankara', 'Çankırı', 'Eskişehir', 'Karaman', 'Kırıkkale', 'Kırşehir', 'Konya', 'Nevşehir', 'Niğde', 'Sivas', 'Yozgat', 'Kayseri'],
    'Ege': ['İzmir', 'Manisa', 'Aydın', 'Denizli', 'Kütahya', 'Afyonkarahisar', 'Uşak', 'Muğla'],
    'Akdeniz': ['Adana', 'Osmaniye', 'Antalya', 'Burdur', 'Hatay', 'Isparta', 'Mersin', 'Kahramanmaraş'],
    'Karadeniz': ['Rize', 'Trabzon', 'Artvin', 'Sinop', 'Tokat', 'Çorum', 'Amasya', 'Samsun', 'Zonguldak', 'Bolu', 'Düzce', 'Karabük', 'Bartın', 'Kastamonu', 'Bayburt', 'Giresun', 'Gümüşhane', 'Ordu'],
    'Doğu Anadolu': ['Ağrı', 'Ardahan', 'Bingöl', 'Bitlis', 'Elazığ', 'Erzincan', 'Erzurum', 'Hakkari', 'Iğdır', 'Kars', 'Malatya', 'Muş', 'Tunceli', 'Van', 'Şırnak'],
    'Güneydoğu Anadolu': ['Adıyaman', 'Batman', 'Diyarbakır', 'Gaziantep', 'Kilis', 'Mardin', 'Siirt', 'Şanlıurfa']
}

def get_region(city):
    for region, cities in geographical_regions.items():
        if city in cities:
            return region
    return None

merged_df['GeographicalRegion'] = merged_df['City'].apply(get_region)

print(merged_df[['City', 'GeographicalRegion']].tail(12))
print(merged_df.dtypes)
```

### Encoding the Geographical  Region 

```python
GeographicalRegion_encoded = pd.get_dummies(merged_df['GeographicalRegion'], prefix='GeographicalRegion')

GeographicalRegion_encoded = GeographicalRegion_encoded.astype(int)

merged_df = pd.concat([merged_df, GeographicalRegion_encoded], axis=1)

merged_df.drop('City', axis=1, inplace=True)
merged_df.drop('GeographicalRegion', axis=1, inplace=True)

print(merged_df.tail(2))
print(merged_df.dtypes)
```
